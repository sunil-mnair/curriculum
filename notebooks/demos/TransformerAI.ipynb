```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch
```

```python
# Load the pre-trained model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
```

```python
# Input text
input_text = "Cats are"

# Encode input text
input_ids = tokenizer.encode(input_text, return_tensors="pt")

# Generate logits
with torch.no_grad():
    outputs = model(input_ids)
    logits = outputs.logits

# Get logits for the last word
last_word_logits = logits[0, -1, :]

# Apply softmax to get probabilities
probs = torch.softmax(last_word_logits, dim=0)

# Top 10 predictions
top_k = 10
top_k_probs, top_k_indices = torch.topk(probs, top_k)
predicted_words = [tokenizer.decode([idx]).strip() for idx in top_k_indices]

# Temperature scaling example
temperature = 0.7  # Adjust for different prediction dynamics
scaled_logits = last_word_logits / temperature
scaled_probs = torch.softmax(scaled_logits, dim=0)
top_k_scaled_probs, _ = torch.topk(scaled_probs, top_k)

```

```python
# Display results
print("Predictions with probabilities (default temperature):")
for word, prob in zip(predicted_words, top_k_probs):
    print(f"{word}: {prob.item():.4f}")

```

```python
print("\nPredictions with probabilities (scaled by temperature):")
for word, prob in zip(predicted_words, top_k_scaled_probs):
    print(f"{word}: {prob.item():.4f}")
```
